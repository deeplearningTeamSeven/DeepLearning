```python
import os, glob
import json
import pandas as pd
import numpy as np
from PIL import Image
```

```python
path = '/content/drive/MyDrive/project/RetinaNet/food'
data_folder = os.listdir(path)
json_folder = []
img_folder = []

for i in data_folder:
    if 'json' in i:
        json_folder.append(i)
    else :
        img_folder.append(i)
json_folder.sort()
img_folder.sort()
```

```python
path = '/content/drive/MyDrive/project/RetinaNet/food/'
annotation_lst = []

for i in img_folder:
    json_file = glob.glob(os.path.join(path + i + ' json', '*.json'))
    json_file.sort()
    json_file = json_file[80:]

    for j in range(len(json_file)):
        # print(json_file[i])
        with open(json_file[j]) as f:
            
            json_data = json.load(f)
            image = Image.open(path + i + '/' + json_data[0]["Code Name"])
            print('img path: ', path + i + '/' + json_data[0]["Code Name"])

            point = json_data[0]["Point(x,y)"]
            x = float(point.split(',')[0])
            y = float(point.split(',')[1])
            w = float(json_data[0]["W"])
            h = float(json_data[0]["H"])
            img_w, img_h = image.size

            dw = 1./img_w
            dh = 1./img_h

            x = x/dw
            y = y/dh

            w = round(w/dw)                 
            h = round(h/dh)     
            x1 = round((2*x - w)/2)         
            y1 = round((2*y - h)/2)         

            x2 = x1 + w
            y2 = y1 + h

            # img_dir_lst.append()
            anno = path + i + '/' + json_data[0]["Code Name"] + ',' + str(x1) + ',' + str(y1) + ',' + str(x2) + ',' + str(y2) + ',' + i
            annotation_lst.append(anno)
            print('길이: ', len(annotation_lst))
anno

#04647
```

```python
np.savetxt('test.csv', annotation_lst, delimiter=',', fmt='%s')
```

```python
df = pd.DataFrame(annotation_lst)
```
